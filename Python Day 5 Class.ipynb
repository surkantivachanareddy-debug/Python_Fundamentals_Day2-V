```python
Organizing the list datatype
```

```python
cars = ['kia','audi','maruthi','hyundai','volkswagen','bens','suzuki','bmw']

print(cars)
```

```python
req-- i want to organise the list in the alphabetical order
```

```python
A-Z (order i want to maintain......!)
```

```python
two combinations
1. temporary approach----->sorted---->we will be getting the orginal order.
2. permanent approach ---->sort---->changes are applied permanently.
```

```python
print(sorted(cars))   #temporary approach 
```

```python
cars.sort()     #permanent approach

print(cars)
```

```python
# i want to know how many cars are there in the list
```

```python
len(cars)
```

```python
# I want to get the Opposite order of the given list like Z-A
```

```python
cars.reverse()

print(cars)
```

```python
students = ['divya', 'harika','naveen','hafsa','karthik','vicky']

print(students)
```

```python
Note: Last value is always exclusive, to include the last value we have to increament the index by +1
```

```python
# req--- I want to get divya and harika in the slice?
```

```python
print(students[0:1])  # for first element
```

```python
print (students[0:2])  # for first two elements
```

```python
print(students[2:4])  # for 2nd and 3rd index value
```

```python
# for every 2nd person of the list
```

```python
print(students[0:6:2])  #for every 2nd element
```

```python
print(students[0:6:1])  #whole list
```

```python
print(students[0:6:3])  #for every 3rd element
```
